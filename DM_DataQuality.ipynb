# EJERCICIO DATA QUALITY


**Objetivo:** Evaluar y mejorar la calidad de los datos utilizando las dimensiones: Completeness, Uniqueness, Validity, Timeliness, Accuracy y Consistency.

**Instrucciones:**

Descarga el siguiente archivo de datos (data_quality_dataset.csv) o genera los datos directamente en Python.
Usa Python para realizar las siguientes evaluaciones y mejoras:
* Identificar duplicados (Uniqueness).
* Verificar si hay valores faltantes (Completeness).
* Comprobar si los valores cumplen las reglas esperadas (Validity).
* Revisar si los datos están actualizados (Timeliness).
* Evaluar la precisión de ciertos registros (Accuracy).
* Confirmar la consistencia de los datos (Consistency).

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from google.colab import files
```

```python
archivo_csv = "data_quality_dataset_100.csv"
df = pd.read_csv(archivo_csv)
print(df)
```

```python
# 2. Uniqueness: Identificar duplicados
print("Registros duplicados:")
#print(df[df.duplicated(subset=['Name'])])
df["Uniqueness"] = ~df.duplicated(subset=['Name'], keep= False)
print(df)
#print(df[df.duplicated()])
```

```python
# 1. Completeness: Verificar valores faltantes

print("Valores faltantes por columna:")
print(df.isnull().sum())
df["Completeness"] = ~df.isnull().any(axis=1)
print(df)
```

```python
# 3. Validity: Comprobar reglas
# - Emails válidos
df["Valid_Email"] = df["Email"].str.contains(r"^[\w\.-]+@[\w\.-]+\.\w+$", na=False)
# - Ingresos positivos
df["Valid_Income"] = df["Income"] > 0

print("\nValidez de los datos:")
print(df[["Email", "Valid_Email", "Income", "Valid_Income"]])
```

```python
# 4. Timeliness: Fechas de ingreso no deben ser futuras
df["Join_Date"] = pd.to_datetime(df["Join_Date"])
df["Timely"] = df["Join_Date"] <= datetime.now()

print("\nRegistros con fechas inválidas:")
print(df[~df["Timely"]])
```

```python
# 5. Accuracy: Verificar ingresos (ejemplo: rango esperado de ingresos entre $0 y $100,000)
df["Accurate_Income"] = df["Income"].between(0, 100000)

print("\nRegistros con ingresos fuera de rango:")
print(df[~df["Accurate_Income"]])
```

```python
# Crear una nueva columna 'Errores' que cuente cuántas evaluaciones son FALSAS
df["Errores"] = ((~df["Uniqueness"]).astype(int) + (~df["Completeness"]).astype(int) + (~df["Valid_Email"]).astype(int) + (~df["Valid_Income"]).astype(int) + (~df["Timely"]).astype(int) + (~df["Accurate_Income"]).astype(int))
# Marcar registros como 'DATO ERRÓNEO' si tienen 2 o más errores
df["Consistency"] = df["Errores"].apply(lambda x: "D.ERRONEO" if x >= 2 else "D.VALIDO")

# Mostrar registros con inconsistencias detectadas
print("\nRegistros con errores en la consistencia de datos:")
print(df[df["Consistency"] == "DATO ERRONEO"])
```

```python
df.to_csv("data_quality_dataset.csv", index=False)
from google.colab import files

# Descargar el archivo guardado
files.download("data_quality_dataset.csv")
```
